```python
#Linear Regression

class Lin_Reg_model():

 # initiating the parameters (learning rate & no. of iterations)
  def __init__(self,learning_rate,no_of_iterations):

    self.learning_rate = learning_rate
    self.no_of_iterations =no_of_iterations


  def fit(self,X,Y):

    #number of traninig examples and number of features
    self.m,self.n = X.shape  #number of rows and columns

    #initiating weight and bias value
    self.w = np.zeros(self.n)
    self.b = 0
    self.X = X
    self.Y = Y


    #implementing gradient descent

    for i in range(self.no_of_iterations):
      self.update_weights()


  def update_weights(self,):

    Y_prediction = self.predict(self.X)

    #calculate gradients
    dw = - (2 * (self.X.T).dot(self.Y - Y_prediction)) / self.m
    db = - 2 * np.sum(self.Y - Y_prediction)/self.m

    #updating the weights
    self.w = self.w - self.learning_rate*dw
    self.b = self.b - self.learning_rate*db





  def predict(self,X):

    return X.dot(self.w) + self.b


```

```python
# importing numpy library
import numpy as np
```

```python
class Logistic_Regression():

  # declaring learning rate and the number of iterations(Hyperparameters)
  def __init__(self, learning_rate, no_of_iterations):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations

  #fit function to train the model
  def fit(self,X,Y):

    self.m, self.n = X.shape  #no: of rows(datapoints) ->m and no: of features(columns) ->n in dataset

    #initiating weight and bias values
    self.w = np.zeros(self.n)
    self.b = 0
    self.X = X
    self.Y = Y

    #implementing gradient descent

    for i in range(self.no_of_iterations):
      self.update_weights()


  def update_weights(self):

    #Y^ formula(sigmoid function)

    Y_hat = 1 / (1 + np.exp(-(self.X.dot(self.w) + self.b)))   #wx + b

    #derivatives

    dw = (1/self.m)*np.dot(self.X.T, (Y_hat - self.Y))
    db = (1/self.m)*np.sum(Y_hat - self.Y)

    #updating the weights & bias using gradient descent

    self.w = self.w - self.learning_rate * dw
    self.b = self.b - self.learning_rate * db


  # Sigmoid equation and decision boundry
  def predict(self,X):

    Y_pred = 1 / (1 + np.exp(-(X.dot(self.w) + self.b)))
    Y_pred = np.where( Y_pred > 0.5, 1, 0)
    return Y_pred
```

#SVM Classifier



Equation of the Hyperplane:

y = wx - b




Gradient Descent:

Gradient Descent is an optimization algorithm used for minimizing the loss function in various machine learning algorithms. It is used for updating the parameters of the learning model.

w = w - *dw

b = b - *db




Learning Rate:

Learning rate is a tuning parameter in an optimization algorithm that determines the step size at each iteration while moving toward a minimum of a loss function.


```python
# importing numpy library
import numpy as np
```

```python
class SVM_classifier():
  # initiating the hyperparameters
  def __init__(self, learning_rate, no_of_iterations, lambda_parameter):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations
    self.lambda_parameter = lambda_parameter

  def fit(self, X, Y):
    # m -> no. of rows, n -> no. of columns
    self.m, self.n = X.shape

    # initiating the weight value and bias value

    self.w = np.zeros(self.n)
    self.b = 0

    self.X = X
    self.Y = Y

    #implementing Gradient Descent algorithm for Optimization

    for i in range(self.no_of_iterations):
      self.update_weights()



  def update_weights(self):

    # label encoding

      y_label = np.where(self.Y <= 0, -1, 1)

    # gradients (dw,db)
      for index, x_i in enumerate(self.X):
        condition = y_label[index] * (np.dot(x_i, self.w) - self.b) >= 1
        if (condition == True):
          dw = 2 * self.lambda_parameter * self.w
          db = 0
        else:
          dw = 2 * self.lambda_parameter * self.w - np.dot(x_i, y_label[index])
          db = y_label[index]

        self.w = self.w - self.learning_rate * dw
        self.b = self.b - self.learning_rate * db


      #predict the label for a given input value

  def predict(self, X):
        output = np.dot(X, self.w) - self.b
        predicted_labels = np.sign(output)
        y_hat = np.where(predicted_labels <= -1, 0, 1)

        return y_hat





```

```python
# =========================
# FULL SVM CODE (ONE CELL)
# =========================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score as acc_score   # renamed to avoid conflict


# -------------------------
# SVM CLASS FROM SCRATCH
# -------------------------
class SVM_classifier():

    def __init__(self, learning_rate, no_of_iterations, lambda_parameter):
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations
        self.lambda_parameter = lambda_parameter
        self.accuracy_history = []

    def fit(self, X, Y, X_val=None, Y_val=None):
        self.m, self.n = X.shape

        self.w = np.zeros(self.n)
        self.b = 0

        self.X = X
        self.Y = Y

        # label encoding (-1, +1)
        self.y_label = np.where(self.Y <= 0, -1, 1)

        # Gradient Descent
        for _ in range(self.no_of_iterations):
            self.update_weights()

            if X_val is not None:
                y_pred = self.predict(X_val)
                acc = acc_score(Y_val, y_pred)
                self.accuracy_history.append(acc)

    def update_weights(self):
        for index, x_i in enumerate(self.X):
            condition = self.y_label[index] * (np.dot(x_i, self.w) - self.b) >= 1

            if condition:
                dw = 2 * self.lambda_parameter * self.w
                db = 0
            else:
                dw = 2 * self.lambda_parameter * self.w - x_i * self.y_label[index]
                db = self.y_label[index]

            self.w -= self.learning_rate * dw
            self.b -= self.learning_rate * db

    def predict(self, X):
        linear_output = np.dot(X, self.w) - self.b
        predicted_labels = np.sign(linear_output)
        return np.where(predicted_labels <= -1, 0, 1)


# -------------------------
# LOAD DATA
# -------------------------
diabetes_data = pd.read_csv(
    "/content/drive/MyDrive/Colab Notebooks/MachineLearningCourseWithPython/data/diabetes.csv"
)

X = diabetes_data.drop(columns='Outcome', axis=1)
Y = diabetes_data['Outcome']


# -------------------------
# TRAIN-TEST SPLIT
# -------------------------
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, stratify=Y, random_state=2
)


# -------------------------
# FEATURE SCALING
# -------------------------
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


# -------------------------
# TRAIN MODEL
# -------------------------
classifier = SVM_classifier(
    learning_rate=0.001,
    no_of_iterations=1000,
    lambda_parameter=0.01
)

classifier.fit(X_train, Y_train, X_test, Y_test)


# -------------------------
# ACCURACY SCORES
# -------------------------
train_accuracy = acc_score(Y_train, classifier.predict(X_train))
test_accuracy = acc_score(Y_test, classifier.predict(X_test))

print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)


# -------------------------
# PLOT ACCURACY HISTORY
# -------------------------
plt.figure(figsize=(8, 5))
plt.plot(classifier.accuracy_history)
plt.xlabel("Iterations")
plt.ylabel("Validation Accuracy")
plt.title("SVM Accuracy vs Iterations")
plt.grid(True)
plt.show()
```

```python
import numpy as np

# creating a class for Lasso Regression

class Lasso_Regression():

  #initiating the hyperparameters
  def __init__(self, learning_rate, no_of_iterations, lambda_parameter):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations
    self.lambda_parameter = lambda_parameter


  # fitting the dataset to the Lasso Regression model
  def fit(self, X, Y):

    # m --> number of Data points --> number of rows
    # n --> number of input features --> number of columns
    self.m, self.n = X.shape

    self.w = np.zeros(self.n)

    self.b = 0

    self.X = X

    self.Y = Y

    # implementing Gradient Descent algorithm for Optimization

    for i in range(self.no_of_iterations):
      self.upadte_weights()


  # function for updating the weight & bias value
  def upadte_weights(self):

    # linear equation of the model
    Y_prediction = self.predict(self.X)

    # gradients (dw, db)

    # gradient for weight
    dw = np.zeros(self.n)

    for i in range(self.n):

      if self.w[i]>0:

        dw[i] = (-(2*(self.X[:,i]).dot(self.Y - Y_prediction)) + self.lambda_parameter) / self.m

      else :

        dw[i] = (-(2*(self.X[:,i]).dot(self.Y - Y_prediction)) - self.lambda_parameter) / self.m


    # gradient for bias
    db = - 2 * np.sum(self.Y - Y_prediction) / self.m


    # updating the weights & bias

    self.w = self.w - self.learning_rate*dw
    self.b = self.b - self.learning_rate*db




  # Predicting the Target variable
  def predict(self,X):

    return X.dot(self.w) + self.b
```

```python
import numpy as np
from collections import Counter
import heapq

class KNN_Classifier:

    def __init__(self, distance_metric='euclidean'):
        """
        Initialize the KNN classifier.
        distance_metric: 'euclidean' or 'manhattan'
        """
        self.distance_metric = distance_metric

    def get_distance(self, training_data_point, test_data_point):
        """
        Compute the distance between two data points.
        Assumes the last element of training_data_point is the label.
        """
        x_train = np.array(training_data_point[:-1])
        x_test = np.array(test_data_point[:-1])

        if self.distance_metric == 'euclidean':
            return np.sqrt(np.sum((x_train - x_test) ** 2))
        elif self.distance_metric == 'manhattan':
            return np.sum(np.abs(x_train - x_test))
        else:
            raise ValueError("Unsupported distance metric. Choose 'euclidean' or 'manhattan'.")

    def nearest_neighbors(self, X_train, test_data, k):
        """
        Find the k nearest neighbors for the given test_data.
        """
        distance_list = [(train_point, self.get_distance(train_point, test_data))
                         for train_point in X_train]

        # Efficiently get k nearest neighbors without sorting the entire list
        k_neighbors = heapq.nsmallest(k, distance_list, key=lambda x: x[1])
        return [neighbor[0] for neighbor in k_neighbors]

    def predict(self, X_train, test_data, k):
        """
        Predict the class label for a single test data point.
        """
        neighbors = self.nearest_neighbors(X_train, test_data, k)
        labels = [data[-1] for data in neighbors]
        predicted_class = Counter(labels).most_common(1)[0][0]  # Handles ties safely
        return predicted_class

    def predict_batch(self, X_train, X_test, k):
        """
        Predict classes for multiple test data points.
        """
        return [self.predict(X_train, test_point, k) for test_point in X_test]
```

```python
import pandas as pd
import numpy as np
from collections import Counter
import heapq

# --- KNN CLASS DEFINITION ---
class KNN_Classifier:
    def __init__(self, distance_metric='euclidean'):
        self.distance_metric = distance_metric

    def get_distance(self, training_data_point, test_data_point):
        """
        Compute distance between training point (includes label) and test point (features only)
        """
        x_train = np.array(training_data_point[:-1])  # drop label
        x_test = np.array(test_data_point)            # keep all features

        if self.distance_metric == 'euclidean':
            return np.sqrt(np.sum((x_train - x_test) ** 2))
        elif self.distance_metric == 'manhattan':
            return np.sum(np.abs(x_train - x_test))
        else:
            raise ValueError("Unsupported distance metric. Choose 'euclidean' or 'manhattan'.")

    def nearest_neighbors(self, X_train, test_data, k):
        distance_list = [(train_point, self.get_distance(train_point, test_data))
                         for train_point in X_train]
        k_neighbors = heapq.nsmallest(k, distance_list, key=lambda x: x[1])
        return [neighbor[0] for neighbor in k_neighbors]

    def predict(self, X_train, test_data, k):
        neighbors = self.nearest_neighbors(X_train, test_data, k)
        labels = [data[-1] for data in neighbors]
        return Counter(labels).most_common(1)[0][0]

    def predict_batch(self, X_train, X_test, k):
        return [self.predict(X_train, test_point, k) for test_point in X_test]


```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

import warnings
warnings.filterwarnings("ignore")
```

```python
# Load dataset
heart_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MachineLearningCourseWithPython/data/heart.csv')

# Features and target
X = heart_data.drop(columns='target')
y = heart_data['target']

print("Dataset shape:", heart_data.shape)
heart_data.head()
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Random Forest": RandomForestClassifier(random_state=42)
}
```

```python
def compare_models_cv(models, X, y, cv=5):
    results = []

    for name, model in models.items():

        # Scale only when required
        if name in ["Logistic Regression", "SVM", "KNN"]:
            pipeline = Pipeline([
                ("scaler", StandardScaler()),
                ("model", model)
            ])
        else:
            pipeline = model

        scores = cross_val_score(
            pipeline,
            X,
            y,
            cv=cv,
            scoring="accuracy"
        )

        results.append({
            "Model": name,
            "Mean Accuracy (%)": round(scores.mean() * 100, 2),
            "Std Dev": round(scores.std(), 4)
        })

    return pd.DataFrame(results)
```

```python
baseline_results = compare_models_cv(models, X, y)
baseline_results
```

```python
param_grids = {
    "Logistic Regression": {
        "model__C": [0.1, 1, 5, 10]
    },
    "SVM": {
        "model__kernel": ["linear", "rbf"],
        "model__C": [1, 5, 10]
    },
    "KNN": {
        "model__n_neighbors": [3, 5, 7, 9]
    },
    "Random Forest": {
        "n_estimators": [50, 100, 200],
        "max_depth": [None, 5, 10]
    }
}
```

```python
def model_selection_gridsearch(models, param_grids, X, y, cv=5):

    results = []

    for name, model in models.items():
        print(f"Running GridSearchCV for {name}...")

        if name in ["Logistic Regression", "SVM", "KNN"]:
            pipeline = Pipeline([
                ("scaler", StandardScaler()),
                ("model", model)
            ])
            params = param_grids[name]
        else:
            pipeline = model
            params = param_grids[name]

        grid = GridSearchCV(
            pipeline,
            params,
            cv=cv,
            scoring="accuracy",
            n_jobs=-1
        )

        grid.fit(X, y)

        results.append({
            "Model": name,
            "Best CV Accuracy (%)": round(grid.best_score_ * 100, 2),
            "Best Parameters": grid.best_params_
        })

    return pd.DataFrame(results)
```

```python
gridsearch_results = model_selection_gridsearch(
    models,
    param_grids,
    X,
    y
)

gridsearch_results
```

```python
best_model_row = gridsearch_results.sort_values(
    by="Best CV Accuracy (%)",
    ascending=False
).iloc[0]

best_model_row
```

```python
# Identify best model
best_row = gridsearch_results.sort_values(
    by="Best CV Accuracy (%)",
    ascending=False
).iloc[0]

best_model_name = best_row["Model"]
best_params = best_row["Best Parameters"]

print("Best Model:", best_model_name)
print("Best Parameters:", best_params)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    roc_auc_score
)

# Rebuild model correctly
if best_model_name in ["Logistic Regression", "SVM", "KNN"]:
    best_model = Pipeline([
        ("scaler", StandardScaler()),
        ("model", models[best_model_name])
    ])
    best_model.set_params(**best_params)
else:
    best_model = models[best_model_name]
    best_model.set_params(**best_params)

best_model
```

```python
from sklearn.model_selection import StratifiedKFold

def evaluate_model_cv(model, X, y, cv=5):

    skf = StratifiedKFold(n_splits=cv, shuffle=True, random_state=42)

    accuracy = cross_val_score(model, X, y, cv=skf, scoring="accuracy")
    precision = cross_val_score(model, X, y, cv=skf, scoring="precision")
    recall = cross_val_score(model, X, y, cv=skf, scoring="recall")
    roc_auc = cross_val_score(model, X, y, cv=skf, scoring="roc_auc")

    return {
        "Accuracy (%)": round(accuracy.mean() * 100, 2),
        "Precision": round(precision.mean(), 3),
        "Recall": round(recall.mean(), 3),
        "ROC-AUC": round(roc_auc.mean(), 3)
    }
```

Precision is the ratio of number of True Positive to the total number of Predicted Positive. It measures, out of the total predicted positive, how many are actually positive.

```python
from sklearn.metrics import precision_score
```

```python
# precision for training data predictions
precision_train = precision_score(Y_train, X_train_prediction)
print('Training data Precision =', precision_train)
```

Recall is the ratio of number of True Positive to the total number of Actual Positive. It measures, out of the total actual positive, how many are predicted as True Positive

```python
from sklearn.metrics import recall_score
```

```python
# recall for training data predictions
recall_train = recall_score(Y_train, X_train_prediction)
print('Training data Recall =', recall_train)
```

```python
# recall for test data predictions
recall_test = recall_score(Y_test, X_test_prediction)
print('Test data Recall =', recall_test)
```

F1 Score is an important evaluation metric for binary classification that combines Precision & Recall. F1 Score is the harmonic mean of Precision & Recall.

```python
from sklearn.metrics import f1_score
```

**Precision, Recall, & F1 Score - function**

```python
def precision_recall_f1_score(true_labels, pred_labels):

  precision_value = precision_score(true_labels, pred_labels)
  recall_value = recall_score(true_labels, pred_labels)
  f1_score_value = f1_score(true_labels, pred_labels)

  print('Precision =',precision_value)
  print('Recall =',recall_value)
  print('F1 Score =',f1_score_value)
```

```python
# classification metrics for training data
precision_recall_f1_score(Y_train, X_train_prediction)
```

```python
# classification metrics for test data
precision_recall_f1_score(Y_test, X_test_prediction)
```
