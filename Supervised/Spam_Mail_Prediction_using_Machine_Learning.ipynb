```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df =pd.read_csv("/content/drive/MyDrive/Colab Notebooks/MachineLearningCourseWithPython/data/mail_data.csv")
```

```python
df['Category'] = df['Category'].map({'spam':0,'ham':1})
```

```python


df['Category'].value_counts()


```

```python
X=df['Message']
y=df['Category']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
print(y_train)
```

```python
#feature extraction -- transform text data to feauture vectors

feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
# print(X_test_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features,y_train)
```

```python
#evaluate the model

train_prediction = model.predict(X_train_features)
train_accuracy = accuracy_score(train_prediction,y_train)
print("Accuracy on training data : ",train_accuracy)
```

```python
test_prediction = model.predict(X_test_features)
test_accuracy = accuracy_score(test_prediction,y_test)
print("Accuracy on test data : ",test_accuracy)
```

```python
#inference system

# input_mail = ["I've been searching for the right words to thank you"]
input_mail = ["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 0845"]
input_mail_features = feature_extraction.transform(input_mail)

prediction = model.predict(input_mail_features)

if prediction[0]==1:
  print("Ham mail")
else:
  print("Spam mail")
```
