```python
# model = Sequential()

# model.add(Conv2D(filters=32,
#                  kernel_size=(4,4),
#                  strides=(1,1),
#                  padding='valid',
#                  input_shape=(28,28,1),
#                  activation='relu'))
# model.add(MaxPool2D(pool_size=(2,2)))

# model.add(Flatten())

# model.add(Dense(128,activation='relu'))
# #Output layer
# model.add(Dense(10,activation='softmax'))


# model.compile(loss='categorical_crossentropy',
#               optimizer='adam',
#               metrics=['accuracy'])

from keras.models import Sequential
from keras.layers import Input, Conv2D, MaxPool2D, Flatten, Dense

model = Sequential([
    Input(shape=(28, 28, 1)),

    Conv2D(
        filters=32,
        kernel_size=(4, 4),
        strides=(1, 1),
        padding='valid',
        activation='relu'
    ),

    MaxPool2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax') #output layer softmax --MULTI Class
])

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=1)
```

```python
model.fit(x_train,y_cat_train,epochs=10,validation_data=(x_test,y_cat_test),callbacks=[early_stop])
```

```python
metrics =pd.DataFrame(model.history.history)
```

```python
metrics[['loss','val_loss']].plot()
```

```python
metrics[['accuracy','val_accuracy']].plot()
```

```python
model.metrics_names
```

```python
model.evaluate(x_test,y_cat_test,verbose=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def show_random_incorrect_predictions(
    X_test,
    y_test,
    predictions,
    class_names=None,
    n_samples=9
):
    y_pred = predictions.argmax(axis=1)
    incorrect_indices = np.where(y_pred != y_test)[0]

    if len(incorrect_indices) == 0:
        print("No incorrect predictions found.")
        return

    n_samples = min(n_samples, len(incorrect_indices))
    selected_indices = np.random.choice(
        incorrect_indices, size=n_samples, replace=False
    )

    cols = int(np.ceil(np.sqrt(n_samples)))
    rows = int(np.ceil(n_samples / cols))

    plt.figure(figsize=(cols * 4, rows * 4))

    for i, idx in enumerate(selected_indices, 1):
        plt.subplot(rows, cols, i)
        plt.imshow(X_test[idx], cmap="gray" if X_test[idx].ndim == 2 else None)
        plt.axis("off")

        true_label = y_test[idx]
        pred_label = y_pred[idx]

        title = (
            f"True: {class_names[true_label]}\nPred: {class_names[pred_label]}"
            if class_names else
            f"True: {true_label}, Pred: {pred_label}"
        )

        plt.title(title, color="red")

    plt.tight_layout()
    plt.show()
```

```python
show_random_incorrect_predictions(
    X_test,
    y_test,
    predictions,
    # class_names=class_names,
    n_samples=2,
    # seed=42
)
```
